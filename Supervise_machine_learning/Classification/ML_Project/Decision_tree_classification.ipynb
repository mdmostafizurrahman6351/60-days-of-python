```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import accuracy_score, classification_report
```

```python

# data load
url = "https://raw.githubusercontent.com/mdmostafizurrahman6351/Maching_learning/refs/heads/main/Supervise_machine_learning/Classification/Dataset/Titanic_%20data.csv"

df = pd.read_csv(url)
print(df.shape)
df.head()
```

```python
# columnes select
x = df[["Pclass", "Sex", "Age", "SibSp", "Parch", "Fare", "Embarked"]]
y = df["Survived"]
print("null value of x: ", x.isnull().sum())
print("null value of Y: ", y.isnull().sum())
```

```python
x.loc[:, "Age"] = x["Age"].fillna(x["Age"].median())
x.loc[:, "Embarked"] = x["Embarked"].fillna(x["Embarked"].mode()[0])
print("null value of x: ", x.isnull().sum())
```

```python
# preprocissing
enc = OneHotEncoder(handle_unknown="ignore", sparse_output=False, drop="first")
enc_data = enc.fit_transform(x[["Sex", "Embarked"]])
enc_cols = enc.get_feature_names_out(["Sex", "Embarked"])
enc_df = pd.DataFrame(enc_data, columns=enc_cols, index=x.index)
x = x.drop(["Sex", "Embarked"], axis=1)
x = pd.concat([x, enc_df], axis=1)
print("x data shape: ", x.shape)
x.head()
```

```python
# train test separate
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.25, random_state=42)

# model create and fit
model = DecisionTreeClassifier()
model.fit(x_train, y_train)

# prediction
y_pred = model.predict(x_test)

# score
print("model score: ", model.score(x_test, y_test))
```

```python
# model evulation
print("Classification score: ", accuracy_score(y_test, y_pred))
print("Classification report: \n", classification_report(y_test, y_pred))
```
