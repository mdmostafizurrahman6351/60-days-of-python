```python
import numpy as np
import pandas as pd
from sklearn.model_selection import KFold
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
url = "https://raw.githubusercontent.com/mdmostafizurrahman6351/Maching_learning/refs/heads/main/Supervise_machine_learning/LinearRegression/Dataset/insurance.csv"

df = pd.read_csv(url)
print(df.shape)
df.head()
```

```python
df.isnull().sum()
```

```python
x = df.drop("charges", axis=1)
y = df["charges"]

```

```python
num_col = x.select_dtypes(include=np.number).columns
cat_col = x.select_dtypes(exclude=np.number).columns
```

```python
pipeline = Pipeline(
    steps=[
        ("scaler", StandardScaler()),
        ("encoder", OneHotEncoder())
    ]
)

preprocessing = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), num_col),
        ("cat", OneHotEncoder(drop="first"), cat_col)
    ]
)

model = Pipeline(
    steps=[
        ("preprocessor", preprocessing),
        ("regressor", LinearRegression())
    ]
)

model.fit(x, y)

# Create a DataFrame for prediction, ensuring column names and correct data types for categorical features
new_data = pd.DataFrame([[19, 'female', 27.900, 0, 'yes', 'southwest']], columns=x.columns)
model.predict(new_data)
```

```python
# model save
import joblib
joblib.dump(model, "Regression.pkl")
```

```python
# model open and predict
model = joblib.load("Regression.pkl")
model.predict(new_data)
```
